```python
%matplotlib inline


import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
```

```python
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
mouse_metadata.head()
```

```python
study_results.head()
```

```python
combined_df = pd.merge(mouse_metadata, study_results, how="inner", on="Mouse ID")
combined_df
```

```python
mouse_count=len(combined_df["Mouse ID"].unique())
mouse_count
```

```python
#duplicate = df[df.duplicated(['Name', 'Age'])]
duplicate_rows = combined_df[combined_df.duplicated(['Mouse ID','Timepoint'])]
duplicate_rows
```

```python
clean_df=combined_df.groupby(['Mouse ID','Timepoint']).agg({'Mouse ID':'count'}).sort_values("Timepoint",ascending=False)
clean_df
```

```python
combined_df = combined_df.drop(combined_df[combined_df["Mouse ID"] == "g989"].index)
combined_df 
```

```python
mouse_count=len(combined_df["Mouse ID"].unique())
mouse_count
```

```python
mean = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dv = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dv, "SEM": sem})
summary_df
```

```python
agg_df = combined_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ['mean', 'median', 'var', 'std', 'sem']})
agg_df
```

```python
drug_data=combined_df.groupby("Drug Regimen").agg({"Mouse ID":"count"})
drug_data.plot(kind="bar",figsize=(10,3))
plt.show()
plt.tight_layout()
```

```python
timepoint_count = combined_df.groupby(["Drug Regimen"]).agg({"Timepoint": "count"})
```

```python
x_axis = combined_df["Drug Regimen"].unique()
plt.bar(x_axis, timepoint_count["Timepoint"])
plt.xlabel("Drug Regimen")
plt.xticks(rotation=45)
plt.ylabel("Timepoint count")
plt.title("Drug Regiment Total Measurements")
plt.show()
```

```python
drug_data=combined_df.groupby("Sex").nunique()['Mouse ID']
drug_data.plot(kind="pie")
plt.show()
```
